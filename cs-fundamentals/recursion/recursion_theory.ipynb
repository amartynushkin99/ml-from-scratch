# Рекурсия

<добавить описание того, что такое рекурсия>

```python
def count_factorial(n, depth=0):
    """ Рекурсивно вычисляет факториал числа """
    
    indent = "  " * depth  # отступ для текущего уровня
    print(f"{indent}→ Вызов factorial({n})")
    
    # Условие выхода
    if n == 0:
        print(f"{indent}  ← Условие выхода: return 1")
        return 1
    
    # Рекурсивный шаг
    result = n * count_factorial(n - 1, depth + 1)
    print(f"{indent}  ← Возврат {n} * factorial({n-1}) = {result}")
    
    return result
```

```python
count_factorial(4)
```
